### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
from collections import Counter

# Split comma-separated strings into lists and count frequencies
langs_have = Counter([lang.strip() for sublist in df['LanguageHaveWorkedWith'].dropna() for lang in sublist.split(';')])
langs_want = Counter([lang.strip() for sublist in df['LanguageWantToWorkWith'].dropna() for lang in sublist.split(';')])

# Convert to DataFrame
lang_df = pd.DataFrame({
    'HaveWorkedWith': pd.Series(langs_have),
    'WantToWorkWith': pd.Series(langs_want)
}).fillna(0).astype(int)

# Plot grouped bar chart of top 10
top_langs = lang_df.sum(axis=1).sort_values(ascending=False).head(10).index
lang_df.loc[top_langs].plot(kind='bar', figsize=(10,6))
plt.title("Top Programming Languages: Worked With vs Want to Work With")
plt.xlabel("Programming Language")
plt.ylabel("Number of Developers")
plt.xticks(rotation=45)
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# Find top 5 countries by response count
top_countries = df['Country'].value_counts().head().index

# Filter the dataset
filtered_df = df[df['Country'].isin(top_countries)]

plt.figure(figsize=(12,6))
sns.countplot(data=filtered_df, x='RemoteWork', hue='Country')
plt.title("Remote Work Distribution by Top 5 Countries")
plt.xticks(rotation=45)
plt.legend(title="Country", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
